```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/DL/Assignment/Dataset/playground-series-s3e17/train.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["id","Product ID"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Machine failure"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Type"]=le.fit_transform(df["Type"])
```

```python
df.head()
```

```python
x=df.drop(columns="Machine failure",axis=1)
x.head()
```

```python
y=df["Machine failure"]
y[0:5]
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
#ANN CLASSIFICATION
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
#adding input layer
model.add(Dense(11,activation="relu"))
#adding hidden layers
model.add(Dense(1600,activation="relu"))
model.add(Dense(200,activation="relu"))
#adding output layer
model.add(Dense(2,activation="softmax"))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=10,batch_size=5,validation_data=(xtest,ytest))
```

```python
ypred=model.predict(xtest)
```

```python
ypred=model.predict([[1,300.6,309.6,1596,36.1,140,0,0,0,0,0]])
ypred = np.argmax(ypred)
output = [0,1]
output[ypred]
```
